```python
import numpy as np 
```

```python
#indexing = get a specific value 
#slicing = get a specific set of value -> start,stop,step,fancy-indexing,boolean-indexing  
```

```python
arr = np.arange(1,10)


print(f"arr: {arr}")
print(f"indexing: {arr[5]}")
print(f"slicing: {arr[2:5]}") #give 2,3,4
print(f"start: {arr[:5]}") # give 0,2,3,4
print(f"steps: {arr[2:5:2]}")
print(f"stop: {arr[2::2]}")
print(f"reverse: {arr[::-1]}") #reverse the array 
```

```python
#fancy indexing - selecting multiple elements at once which are not sequential 
arr = np.arange(1,10)
print(f"arr: {arr}")
print(f"fancy indexing: {arr[[2,3,6]]}")
 
```

```python
#filtering(boolean masking) = get data based on conditions 
arr = np.arange(1,10)
print(f"arr: {arr}")
print(f"boolean masking: {arr[arr > 5]}")
print(f"boolean masking: {arr[arr <= 5]}")

```
