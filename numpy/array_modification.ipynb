```python
import numpy as np 
```

```python
#insert = insert new element into existing one 
#np.insert(array,index,value,axis = None)
arr = np.arange(1,6)
print(arr)
new_arr = np.insert(arr,2,78,axis = None)
print(new_arr)
```

```python
multi = np.array([[1,2,3],[4,5,6]])
print(multi)

new_arr = np.insert(multi,1,[-1,-1,-1],axis = 0)
print(new_arr)
```

```python
#append 
arr1 = np.arange(1,6)
print(arr1)
arr2 = np.arange(6,11)
print(arr2)
print(np.append(arr1,arr2))
```

```python
#concatinate 
arr1 = np.arange(1,6)
print(arr1)
arr2 = np.arange(6,11)
print(arr2)
print(np.concatenate((arr1,arr2),axis = 0))
```

```python
#delete  
# - 1D array index = 0 
# - 2D array index = list at index 0  

arr1 = np.arange(1,6)
print(arr1)
arr2 = np.arange(6,11)
print(arr2)
print(np.delete(arr1,1,axis = 0))
```

```python
#stacking 
#vstack 
#hstack 


arr1 = np.arange(1,6)
print(arr1)
arr2 = np.arange(6,11)
print(arr2)

print(np.vstack((arr1,arr2)))
print(np.hstack((arr1,arr2)))

```

```python
#splitting 

#vsplit()
#hsplit() 


arr1 = np.arange(1,5)
print(arr1)
arr2 = np.arange(6,11)
print(arr2)

print(np.split(arr1,2))
```
