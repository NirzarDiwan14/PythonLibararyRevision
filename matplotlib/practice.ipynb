```python
#PyPlot API 
#unvariate - focusing on single column at a time 
df = pd.DataFrame({"Salary" : [50,60,30,40,10,90,500,40,90,10]  *5 })
```

```python
#line

plt.plot(df['Salary'],color = "purple",marker = "o",linestyle = "--",linewidth = "2")
plt.grid()
plt.show()
```

```python
#histogram 
plt.hist(df['Salary'],bins = 5,color = "red")
plt.show()
```

```python
#boxplot 
plt.boxplot(df['Salary'])
plt.show()
```

```python
#univariate -- categorical data 
import random 
df['dept'] = ["hr", "it", "finance", "it", "testing"] * 10
df['dept']

```

```python
count = df['dept'].value_counts()

plt.pie(count,labels = count.index,autopct = "%1.1f",explode = [0.1,0,0.2,0])
plt.axis("equal")
plt.show()
```

```python
plt.bar(count.index,count,color = ["red","blue","green","purple"])
plt.show()
```

```python
#bivariate = numerical -numerical 
df['age'] = [int(random.random() * 100) for _ in range(len(df))]
df
```

```python
#scatter plot 
plt.scatter(df['Salary'],df['age'],color = "orange")
plt.show()
```
