## 1. Importing Pandas

**What:** Pandas is a Python library for data manipulation and analysis.

**Why:** It provides powerful data structures like DataFrame and Series.

**How:** Import using `import pandas as pd`.

**When:** Whenever you need to work with tabular data.


```
import pandas as pd
```

## 2. Loading Data

**What:** Reading data from CSV files into a DataFrame.

**How:** Use `pd.read_csv()`.

**Where:** Useful for data stored in files.


```
df = pd.read_csv('raw_data.csv')
df
```

## 3. Inspecting Data

**What:** Understanding structure and summary of data.

- `head()` shows first rows
- `tail()` shows last rows
- `info()` gives column types
- `describe()` gives statistics


```
df.head(10)
df.tail(10)
df.info()
```

```
df.describe()
```

## 4. Selecting Columns

**What:** Extract specific columns.

**How:** Use double brackets.


```
df[['order_amount','payment_mode']]
```

## 5. Filtering Data

**What:** Select rows based on conditions.

**How:** Boolean indexing.


```
condition = df[(df['order_amount'] > 20) & (df['payment_mode'] == 'Cash')]
len(condition)
```

## 6. Adding Columns

**What:** Create new columns.

**How:** Assign lists or use `insert()`.


```
df['new_data'] = [i + 2 for i in range(len(df))]
df
```

```
data = [i + 3 for i in range(len(df))]
df.insert(2,'new_data_2',data)
df
```

## 7. Updating Values

**What:** Modify specific cell values.

**How:** Use `.loc[row, col]`.


```
df.loc[0,'payment_mode'] = 'Cash'
df
```

## 8. Removing Columns

**What:** Delete unwanted columns.

**How:** Use `drop()`.


```
df.drop(columns=['new_data_2'], inplace=True)
df
```

## 9. Handling Missing Data

**Definitions:**
- *Missing Values:* Empty or NaN entries.

**Methods:**
- `isnull()`
- `dropna()`
- `fillna()`
- `interpolate()`


```
df.isnull().sum()
```

```
df.dropna(axis=1, inplace=True)
df
```

```
df.fillna(40, inplace=True)
df
```

## 10. Interpolation

**What:** Fill missing values using trends.

**Types:** linear, polynomial, time.


```
df = pd.read_csv('raw_data.csv')
df.interpolate(method='linear', axis=0, inplace=True)
df
```

## 11. Sorting Data

**What:** Arrange rows by column values.


```
df.sort_values(by=['order_date','order_amount'], ascending=[True,True], inplace=True)
df
```

## 12. Aggregation

**What:** Compute statistics.

**Examples:** min, max, mean, std, sum, count.


```
print(df['order_amount'].min())
print(df['order_amount'].max())
print(df['order_amount'].mean())
print(df['order_amount'].std())
print(df['order_amount'].sum())
print(df['order_amount'].count())
```

## 13. Grouping Data

**What:** Group rows and apply functions.


```
df.groupby('payment_mode')['order_amount'].sum()
```

## 14. Merging DataFrames

**What:** Combine DataFrames.

**How:** Use `pd.merge()`.

**Types:** left, right, inner, outer, cross.


```
merged_df = pd.merge(df, df, on='user_id', how='right')
merged_df
```
